## Extract information from the given web site
You will extract the data from the below web site: <br> 


```python
#this url contains the data you need to scrape
url = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBM-DA0321EN-SkillsNetwork/labs/datasets/Programming_Languages.html"
```

The data you need to scrape is the **name of the programming language** and **average annual salary**.<br> It is a good idea to open the url in your web broswer and study the contents of the web page before you start to scrape.


Import the required libraries


```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

Download the webpage at the url


```python
url = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBM-DA0321EN-SkillsNetwork/labs/datasets/Programming_Languages.html"
response = requests.get(url)
```

Create a soup object


```python
soup = BeautifulSoup(response.content, 'html.parser')
```

Scrape the `Language name` and `annual average salary`.


```python
# Scrape the Language name and annual average salary
# D'après l'inspection de la page, les données sont dans un tableau

table = soup.find('table')

languages = []
salaries = []

# Parcourir toutes les lignes du tableau (en commençant par la deuxième pour éviter l'en-tête)
for row in table.find_all('tr')[1:]:  # [1:] pour sauter la première ligne d'en-tête
    columns = row.find_all('td')
    
    if len(columns) >= 2:  # S'assurer qu'il y a au moins 2 colonnes
        language = columns[1].get_text(strip=True)  # La langue est dans la 2ème colonne
        salary = columns[3].get_text(strip=True)    # Le salaire est dans la 4ème colonne
        
        languages.append(language)
        salaries.append(salary)

# Créer un DataFrame avec les données extraites
df = pd.DataFrame({
    'Programming Language': languages,
    'Average Annual Salary': salaries
})
```

Save the scrapped data into a file named *popular-languages.csv*


```python
df.to_csv('popular-languages.csv', index=False)

print("Données extraites et sauvegardées dans 'popular-languages.csv'")
print(f"Nombre de langages extraits: {len(df)}")
print("\nAperçu des données:")
print(df.head())
```
